```python
import sys

verbose = False

def count_syllables(word):
    '''Based off of countsyl (https://github.com/akkana/scripts/blob/master/countsyl)'''
    
    vowels = ['a', 'e', 'i', 'o', 'u']
    oneGang = ['ai','ao','ei','ia','ie','io','iu','ou','ua','ue','ui','uo']
    twoGang = ['ee','oe']

    on_vowel = False
    in_diphthong = False
    minsyl = 0
    maxsyl = 0
    lastchar = ''

    word = word.lower()
    for c in word:
        is_vowel = c in vowels

        if on_vowel == None:
            on_vowel = is_vowel

        if is_vowel:
            if verbose: print("%s is a vowel" % c)
            if not on_vowel:
                # We weren't on a vowel before.
                # Seeing a new vowel bumps the syllable count.
                if verbose: print("new syllable")
                minsyl += 1
                maxsyl += 1
            elif on_vowel and not in_diphthong and c != lastchar:
                # We were already in a vowel.
                # Don't increment anything except the max count,
                # and only do that once per diphthong.
                if verbose: print("%s is a diphthong" % c)
                in_diphthong = True
                # maxsyl += 1
            if lastchar is None and (lastchar + c) in twoGang:
                maxsyl += 1
        elif verbose: print("[consonant]")
        on_vowel = is_vowel
        lastchar = c

    if not minsyl:
        minsyl = 1

    return maxsyl
```

```python
import pandas as pd

syllableTest = pd.read_csv('TangEngNames.csv', header=None, names=['firstName'])
syllableTest.loc[syllableTest['firstName'].str.split().str.len() == 2, 'lastName'] = syllableTest['firstName'].str.split().str[-1]
syllableTest['firstName'] = syllableTest['firstName'].str.split().str[0]
syllableTest.head()
```

```python
def repeatVowel(str):
    '''Returns vowels that sit next to each other in a word.'''

    str = str.lower()
    vowels = ['a', 'e', 'i', 'o', 'u']
    l = len(str)
    concurVowel = ''
  
    # Find any vowel sitting next to another vowel.
    x = str[0] in vowels
    for i in range(1, l): 
        y = str[i] in vowels
        if x and y:
            concurVowel = str[i-1:i+1]
        x = y
    return concurVowel
```

```python
syllableTest['firstSyl'] = syllableTest['firstName'].apply(count_syllables)
syllableTest['lastSyl'] = syllableTest['lastName'].apply(count_syllables)
syllableTest['firstDouble'] = syllableTest['firstName'].apply(repeatVowel)
syllableTest['lastDouble'] = syllableTest['lastName'].apply(repeatVowel)
display(syllableTest)
```

```python
syllableTest.firstDouble.append(syllableTest.lastDouble).sort_values().unique()
```

```python
def multiVowel(word):
    '''Identify words where a vowel appears three or more times.'''
    
    vowels = ['a', 'e', 'i', 'o', 'u']
    
    groupVowel = ''
    cList = ''
    vCount = 0
    for c in word:
        if c in vowels:
            vCount += 1
            cList += c
        elif c not in vowels:
            vCount = 0
            cList = ''
        if vCount > 2:
            groupVowel = cList
    return groupVowel
```

```python
syllableTest['firstMulti'] = syllableTest['firstName'].apply(multiVowel)
syllableTest['lastMulti'] = syllableTest['lastName'].apply(multiVowel)
display(syllableTest)
```

```python
syllableTest.firstMulti.append(syllableTest.lastMulti).sort_values().unique()
```
